# Data Visualization

```python
# Merge train(sales) with store data
train_merge = pd.merge(train, store, on='Store')
train_merge.head()
```

```python
# Open - missing data
train_merge['Open'] = train_merge['Open'].apply(lambda x: 0 if np.isnan(x) else x)

pd.isnull(train_merge).any()
```

```python
train_merge.Date = train_merge.Date.astype(str)
train_merge['Year'] = train_merge.Date.apply(lambda x: x[0:4])
train_merge['Month'] = train_merge.Date.apply(lambda x:x[5:7])
train_merge['Day'] = train_merge.Date.apply(lambda x:x[8:10])
train_merge = train_merge.drop('Date',1)
train_merge['StateHoliday'] = map(lambda x: str(x), train_merge['StateHoliday'])
train_merge['Year'] = map(lambda x: int(x), train_merge['Year'])
train_merge['Month'] = map(lambda x: int(x), train_merge['Month'])
train_merge['Day'] = map(lambda x: int(x), train_merge['Day'])
```
